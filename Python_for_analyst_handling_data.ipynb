```python
import pandas as pd
import random
```

```python
a = [1,2,3,4,56,7,78,88,99]
```

```python
with open('xx.txt', 'w') as f:
  for i in a:
    f.write(str(i))
```

```python
a1 = '\n'.join([str(x) for x in a])
with open('xx1.txt', 'w') as f:
  f.writelines(a1)
```

```python
column_names = ['c'+str(i) for i in range(20)]
print(column_names)
```

```python
a2 = [random.randint(1,200) for x in column_names]
print(a2)
```

```python
column_str = ','.join(column_names)
with open('xx.txt', 'w') as f:
    f.write(column_str+'\n')
```

```python
with open('xx.txt', 'a') as f:
  for i in range(5):
    a2_str = ','.join([str(random.randint(1,200)) for x in column_names])
    f.writelines(a2_str+'\n')
```
